### 6.2 Handling missing data

As we can see in above figure only two attribute have missing value i.e age and tenure.

```python
import numpy as np
null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency'] = data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100
null_frame.transpose()
```

Observations:

- We can observe that following features are found to have missing values along with the proportions:

|Feature|Object Type|Missing Proportion|Solution|
|:--:|:--:|:--:|:--|
|gender|Object|17%|Replace with mode.|
|tenure|Object|0.02%|Replace with mode.|


```python

data['gender'] = data['gender'].replace(np.nan, data['gender'].mode()[0])
data['tenure'] = data['tenure'].replace(np.nan, data['tenure'].mode()[0])

null_frame = pd.DataFrame(index = data.columns.values)
null_frame['Null Frequency'] = data.isnull().sum().values
percent = data.isnull().sum().values/data.shape[0]
null_frame['Missing %age'] = np.round(percent, decimals = 4) * 100
null_frame.transpose()
```

### we handled missing value by replacing with mode.

## Q2 To Which type of user is more active app user or non app user?

```python
labels=['10-20','21-30','31-40','41-50','51-60','61-70','71-80','81-90','91-100','101-110','111-120']
data['age'] = pd.cut(data.age,bins=np.arange(10,121,10),labels=labels,right=True)

sns.barplot(data['age'],data['mobile_likes'])
plt.title("Mobile users")
```

```python
sns.barplot(data['age'],data['www_likes'])
plt.title("Website users")
```

### As you can see from graph web user are more active than app user.
